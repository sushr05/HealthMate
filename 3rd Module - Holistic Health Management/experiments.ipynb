```python
from langchain_community.document_loaders import DirectoryLoader
from langchain_community.document_loaders import PyPDFLoader

loader = DirectoryLoader(
    path="research-papers", glob="*.pdf", show_progress=True, loader_cls=PyPDFLoader
)

data = loader.load()
```

```python
from langchain_text_splitters.sentence_transformers import SentenceTransformersTokenTextSplitter

st_text_splitter = SentenceTransformersTokenTextSplitter(
    model_name="sentence-transformers/all-mpnet-base-v2", 
    chunk_size=100, 
    chunk_overlap=50
)
```

```python
st_chunks = st_text_splitter.create_documents(doc_content, doc_metadata)
```

```python
print("Total number of documents inside chunks:", len(st_chunks))
print()
for i, chunk in enumerate(st_chunks, start=1):
    print(f"Document {i} metadata: {chunk.metadata}")
    print(f"Document {i} chunk (first 100 chars): {chunk.page_content[:100]}")
    if i == 5:
        break
    print("-" * 100)
```

```python
db.add_documents(st_chunks)
```

```python
print(db.get()['documents'][1][:100])
print(db.get()['metadatas'][:5])
```

```python
retriever = db.as_retriever(search_type="similarity", search_kwargs={'k': 5})
```

```python
from langchain_core.prompts import ChatPromptTemplate

PROMPT_TEMPLATE = """
You are a highly knowledgeable assistant specializing in Holistic Health Management.  
Your expertise covers **physical fitness, mental wellness, diet, and lifestyle habits**.  
You provide only **personalized suggestions** and avoid **medical-related queries** (e.g., diseases, symptoms, medications).  

Use the following context to generate responses:  
{context}

User Query:  
{question}

Guidelines:  
- Stick strictly to the context.  
- Provide **specific and actionable** recommendations.  
- **Do NOT** answer medical-related queries.  
- **Do NOT** say "according to the context" or similar phrases.  
- Keep responses **conversational and user-friendly**.  

Begin by asking a personalized question if needed, and then provide a response.
"""

prompt_template = ChatPromptTemplate.from_template(PROMPT_TEMPLATE)
```

```python

from langchain_core.output_parsers import StrOutputParser

output_parser = StrOutputParser()
```

```python
from langchain_core.runnables import RunnablePassthrough

def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)

rag_chain = (
    {"context": retriever | format_docs, "question": RunnablePassthrough()} 
    | prompt_template 
    | chat_model 
    | output_parser
)
```

```python
# Example query invocation
query = "Effects of Overweight on Physical Health ?"
print(rag_chain.invoke(query))
```

```python
from IPython.display import display, Markdown
import textwrap

def to_markdown(text):
    text = text.replace('•', '  *')
    return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

```python
query = "Effects of Overweight on Physical Health ?"
response = rag_chain.invoke(query)
display(to_markdown(response))
```
